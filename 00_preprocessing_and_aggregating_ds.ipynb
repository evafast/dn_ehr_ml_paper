**00_preprocessing_and_aggregating_ds**

output is a .parquet DF aggregating values over time. This creates all the files that are used as inputs for ML. 

Conditions are aggregating over

1) all time available before biopsy (years before biopsy 0 aggregates all)
2) only 1 year before biopsy  
3) aggregate for one year but with different gaps before the biopsy (1 - 5 years).   
4) only 2 years before biopsy  
5) aggregate for two years but with different gaps before the biopsy (1 - 5 years).
6) aggregating for different time periods before biopsy (0, 1, 2 already done above)-  do 3, 4, 5 years.
